```python
import numpy as np
import pandas as pd
```

```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("prasoonkottarathil/polycystic-ovary-syndrome-pcos")

print("Path to dataset files:", path)
```

```python

file_path = f"{path}/PCOS_data_without_infertility.xlsx"

df = pd.read_excel(file_path , sheet_name="Full_new")

df.head()
```

```python
df.isnull().sum()
```

```python
df=df.drop(columns=['Sl. No','Patient File No.','Weight (Kg)','Height(Cm) ','Blood Group','Pulse rate(bpm) ','RR (breaths/min)','Hb(g/dl)','Cycle(R/I)','Cycle length(days)','Marraige Status (Yrs)','Pregnant(Y/N)','No. of aborptions','  I   beta-HCG(mIU/mL)', 'II    beta-HCG(mIU/mL)' ,'Hip(inch)','Waist(inch)','Waist:Hip Ratio','TSH (mIU/L)','PRL(ng/mL)','Vit D3 (ng/mL)','PRG(ng/mL)','BP _Systolic (mmHg)', 'Follicle No. (L)', 'Follicle No. (R)','Avg. F size (L) (mm)', 'Avg. F size (R) (mm)','Endometrium (mm)','Unnamed: 44'])
```

```python
df.head()
```

```python
df.info()
```

```python
df['AMH(ng/mL)'] = pd.to_numeric(df['AMH(ng/mL)'],errors='coerce')
```

```python
df.info()
```

```python
df['AMH(ng/mL)'].fillna(df['AMH(ng/mL)'].median(), inplace=True)
```

```python
df.info()
```

```python
df['Fast food (Y/N)'].fillna(df['Fast food (Y/N)'].mode()[0] , inplace=True)
```

```python
df.info()
```

```python
target=df['PCOS (Y/N)']
df.drop(columns=['PCOS (Y/N)'],inplace=True)
```

```python
from imblearn.over_sampling import BorderlineSMOTE
from collections import Counter
oversample = BorderlineSMOTE()
data, target = oversample.fit_resample(df, target)
```

```python
data.shape
```

```python
data.info()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test, y_train, y_test = train_test_split(data, target, test_size=0.15, random_state=1, stratify = target)
```

```python
import xgboost as xgb
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
xgb_clf = xgb.XGBClassifier(n_estimators=100, learning_rate=0.1, max_depth=5, random_state=42)

# Train the model
xgb_clf.fit(X_train, y_train)
```

```python
y_pred = xgb_clf.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)

print(f"Model Accuracy: {accuracy:.2f}")
```

```python
from sklearn.metrics import confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt

cm = confusion_matrix(y_test, y_pred)
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
plt.show()

```
